```python
from datasets import load_dataset
dataset = load_dataset('sentiment140')
```

```python
import re

# preprocess
def preprocess_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+|@\S+|#\S+", '', text)
    text = re.sub(r'[^\w\s]', '', text)
    return text
dataset = dataset.map(lambda x: {"text": preprocess_text(x['text'])})
```

```python
dataset
```

```python
X_train = [x['text'] for x in dataset['train']]
y_train = [x['sentiment'] for x in dataset['train']]

X_test = [x['text'] for x in dataset['test']]
y_test = [x['sentiment'] for x in dataset['test']]

```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_train_vectorized = vectorizer.fit_transform(X_train)
X_test_vectorized = vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import precision_score, recall_score, f1_score

clf = MultinomialNB()
clf.fit(X_train_vectorized, y_train)
y_pred = clf.predict(X_test_vectorized)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

```python
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
